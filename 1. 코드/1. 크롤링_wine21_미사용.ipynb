## wine21 :가격, 아로마, 타닌 크롤링

```python
import selenium
import os 
from bs4 import BeautifulSoup as bs
from urllib import request as rq
from urllib.request import urlopen
import pandas as pd
import numpy as np
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
shin_df_new = shin_df.copy()
shin_df_new.head()
```

```python
shin_df_new
```

```python
shin_df_new['가격'] = np.nan
shin_df_new['탄닌'] = np.nan
shin_df_new['아로마'] = np.nan
```

```python
shin_df_new
```

```python
path = 'C:/cakd7/프로그램 설치/chromedriver.exe'     #집
url = 'https://www.wine21.com/13_search/wine_list.html'

driver = webdriver.Chrome(path)
driver.get(url)

for idx in range(len(shin_df_new)):

    # 검색어 입력
    search_bar = driver.find_element(By.XPATH, '//*[@id="shKeyword1"]')

    ## 한국어이름 입력
    search_contents = str(shin_df_new.loc[idx, '와인이름']).strip()
    search_bar.send_keys(search_contents)
    search_bar.send_keys(Keys.ENTER)

    time.sleep(2)

#     ## 없으면 영어이름으로 검색
#     search_contents = str(shin_df_new.loc[idx, '영어이름']).strip()
#     search_bar.send_keys(search_contents)
#     search_bar.send_keys(Keys.ENTER)

    
    # 해당 와인 설명 페이지 접속 후 크롤링
    try:
        elements = driver.find_elements(By.XPATH, '//*[@id="wine_list"]/li/div[2]/div[1]/h3/a')
        if len(elements) > 1:
            for element in elements:
                if re.sub('[^ㄱ-ㅣ가-힣]+', '', search_contents) == re.sub('[^ㄱ-ㅣ가-힣]+', '', str(element.text)):
                    element.click()
                    
                    
        elif len(elements) == 1:
            for element in elements: 
                element.click()

        ## 가격
        price = driver.find_element(By.XPATH, '/html/body/section/div[3]/div[2]/div[2]/p[1]/strong').text
        shin_df_new.loc[idx, '가격'] = price
        print(price)

        ## 탄닌
        tannine = driver.find_elements(By.XPATH, '/html/body/section/div[3]/div[2]/div[2]/div[2]/ul/li[4]/div/a')
        tan = []
        for t in tannine:
            tan.append(t.get_attribute('class'))
        shin_df_new.loc[idx, '탄닌'] = tan.count('on')
        print(tan.count('on'))

        ## 아로마
        aroma = driver.find_elements(By.XPATH, '/html/body/section/div[3]/div[2]/div[2]/div[3]/div[1]/div[2]/ul/li/p')

        arom = ''
        for a in aroma:
            arom += str(a.text)+', '

        shin_df_new.loc[idx, '아로마'] = arom
        print(arom)
        
        ## 도수
        dosu = driver.find_element(By.XPATH, '//*[@id="detail"]/div/div/dl[6]/dd').text
        shin_df_new.loc[idx, '도수'] = dosu
        print(dosu)
        
        driver.implicitly_wait(time_to_wait = 7)
        time.sleep(1)
        driver.get(url)

        
    except:
        driver.get(url)
        time.sleep(1)
    
    
    
driver.close()
```
