#### 한국어 버전

````python
# 라벨 이미지 내 텍스트 추출
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'

text = pytesseract.image_to_string('wine_인식_투바인샤도네이.jpg',
                                    lang = 'kor+eng', 
                                    config = '-c preserve_interword_space=1 --psm 4')

import pandas as pd
wine_df = pd.read_csv('wine_final_real.csv')

# 라벨 텍스트에서 텍스트, 숫자로만 이루어진 단어 추출
word_li = re.findall('\w+',text)



# 데이터프레임 내 와인 이름 단어 추출
wine_li = wine_df['와인이름'].tolist()

# 삭제해야하는 특수문자 지정
stopword = ['&',',','‘','’','“','”']


# 정규표현식으로 삭제해야하는 단어 제거
# 띄어쓰기제거 한 와인이름 리스트도 생성
wine_li_delete_stopword = []
wine_li_delete_blank = []

for name in wine_li:
    for s in stopword:
        name = re.sub(s, '', name)
        name2 = re.sub('\s','', name)
    wine_li_delete_stopword.append(name)
    wine_li_delete_blank.append(name2)
    
    
# 데이터프레임 내 와인이름 단어 리스트 중복 제거
wine_str_set = list(set(' '.join(wine_li_delete_stopword).split()))


# 중복제거된 와인이름 단어 리스트 내 존재하면 찾아야할 와인이름 완성하기
wine_name = ''
for w in word_li:
    if w in wine_str_set:
        wine_name += w
print(wine_name)


# 해당 와인 이름과 85% 이상 일치하는 와인 인덱스 불러오기
from difflib import SequenceMatcher

for idx, df_wine_name in enumerate(wine_li_delete_blank):
    if SequenceMatcher(None, df_wine_name, wine_name).ratio() > 0.8:
        display(wine_df.loc[idx])
````

##### 데이터프레임 내 와인이름(국문/영문) 전처리

````python
import pandas as pd
wine_df = pd.read_csv('wine_final_real.csv')

# 데이터프레임 내 와인 이름 단어 추출
wine_li_ko = wine_df['와인이름'].tolist()
wine_li_en = wine_df['영어이름'].tolist()



# 삭제해야하는 단어 지정하기 위해 모든 글자 추출 
all_str_ko = []
for name in wine_li_ko:
    for n in name:
        all_str_ko.append(n)

all_str_en = []
for name in wine_li_en:
    for n in name:
        all_str_en.append(n)


# 삭제해야하는 단어 지정
stopword_ko = ['&',',','‘','’','“','”']
stopword_en = ['\"','&', "\'", '\(', '\)', ',', '-', '[.]', '/','‘','’','“','”']



# 정규표현식으로 삭제해야하는 단어 제거
# 띄어쓰기제거 한 와인이름 리스트도 생성

wine_li_delete_stopword_ko = []
wine_li_delete_blank_ko = []
for name in wine_li_ko:
    for s in stopword_ko:
        name = re.sub(s, '', name)
        name2 = re.sub('\s','', name)
    wine_li_delete_stopword_ko.append(name)
    wine_li_delete_blank_ko.append(name2)
    



wine_li_delete_stopword_en = []
wine_li_delete_blank_en = []
for name in wine_li_en:
    name = name.lower()
    for s in stopword_en:
        name = re.sub(s, '', name)
        name2 = re.sub('\s','', name)
    wine_li_delete_stopword_en.append(name)
    wine_li_delete_blank_en.append(name2)

    
# 데이터프레임 내 와인이름 단어 리스트 중복 제거
wine_str_set_ko = list(set(' '.join(wine_li_delete_stopword_ko).split()))
wine_str_set_en = list(set(' '.join(wine_li_delete_stopword_en).split()))
````

````python
# 한국어인지 영어인지 확인하는 사용자함수 생성
def is_English_Korean(str):
    k_cnt, e_cnt = 0, 0
     
    for s in str:
        # ord: 문자의 유니코드 값을 돌려주는 함수
        if ord('가') <= ord(s) <= ord('힣'):
            k_cnt += 1
        elif ord('a') <= ord(s.lower()) <= ord('z'):
            e_cnt += 1
            
    return "한국어" if k_cnt > e_cnt else "영어"

is_English_Korean('테스트 아아'), is_English_Korean('test_ahah')
````
