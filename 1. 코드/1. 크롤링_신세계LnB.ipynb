```python
# 신세계

elec_df = pd.DataFrame(columns = ['와인이름', '영어이름', '한줄소개', '와인타입', '나라, 와이너리', '포도종', '용량', '곁들일 음식', '당도', '산도', '바디', '와인설명', '수상', '와이너리이름', '와이너리설명'])

name_koli = []
name_enli = []
short_int = []
winetype = []
country_winery = []
grape_li = []
ml_li = []
food_li = []
dang_li = []
acidity_li = []
body_li = []
infor_li = []
awards_li = []
winery_name = []
winery_short = []
winery_info = []

id = np.arange(5, 2107)
ready = []
error = []
for i in tqdm_notebook(id):
    url = "http://www.shinsegae-lnb.com/product/wineView?id=" + str(i)

    # 각페이지 들어가기
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
    wines = req.get(url, headers = headers)
    bs = BeautifulSoup(wines.text, 'lxml')
    
    if bs.find(class_ = 'error-404'):
        pass
    elif bs.select_one('#section2 > div > div.left_box > img')['alt'] == '이미지 준비중입니다':
        ready.append(i)
    else:
        # 한글이름
        name_ko = bs.select_one('#section2 > div > div.right_box > div.box1 > dl > dt').text
        name_koli.append(name_ko)
        # 영어이름
        name_en = bs.select_one('#section2 > div > div.right_box > div.box1 > dl > dd.etit').text
        name_enli.append(name_en)
        # 한줄소개
        shortinfo = bs.select_one('#section2 > div > div.right_box > div.box1 > dl > dd.txt').text
        short_int.append(shortinfo)
        # 와인타입
        winety = bs.select_one('#section2 > div > div.right_box > div.box2 > ul > li.type1 > span:nth-child(2)').text
        winetype.append(winety)
        # 나라, 와이너리
        cou_wi = bs.select_one('#section2 > div > div.right_box > div.box2 > ul > li.type2 > span:nth-child(2)').text
        country_winery.append(cou_wi)
        # 포도종
        try:
            grape = bs.select_one('#section2 > div > div.right_box > div.box2 > ul > li.type3 > span:nth-child(2)').text
            grape_li.append(grape)
        except:
            grape_li.append(' ')
        # 용량
        try:
            ml = bs.select_one('#section2 > div > div.right_box > div.box2 > ul > li.type4 > span:nth-child(2)').text
            ml_li.append(ml)
        except:
            ml_li.append(' ')
        # 음식
        try:
            food = bs.select_one('#section2 > div > div.right_box > div.box2 > ul > li.type5 > span:nth-child(2)').text
            food_li.append(food)
        except:
            food_li.append(' ')
        # 당 
        try:
            dang = bs.select_one('#section2 > div > div.right_box > div.box3 > dl:nth-child(1) > dd > span.on > p.hidden').text
            dang_li.append(dang)
        except:
            dang_li.append(' ')
        try:
        # 산소
            acid = bs.select_one('#section2 > div > div.right_box > div.box3 > dl:nth-child(2) > dd > span.on > p.hidden').text
            acidity_li.append(acid)
        except:
            acidity_li.append(' ')
        # 바디감
        try:
            body = bs.select_one('#section2 > div > div.right_box > div.box3 > dl:nth-child(3) > dd > span.on > p.hidden').text
            body_li.append(body)
        except:
            body_li.append(' ')
        # 정보
        try:
            infor = bs.select_one('#section3 > div > div.left_box > dl > dd').text
            infor_li.append(infor)
        except:
            infor_li.append(' ')
        # 수상
        try:
            award = bs.select_one('#section3 > div > div.right_box > dl > dd > ul').text
            awards_li.append(award)
        except:
            awards_li.append(' ')
        # 와이너리이름
        try:
            winerna = bs.select_one('#section4 > div > div.box2 > dl > dt').text
            winery_name.append(winerna)
        except:
            winery_name.append(' ')
        # 와이너리특징
        try:
            winersho = bs.select_one('#section4 > div > div.box2 > dl > dd:nth-child(2)').text
            winery_short.append(winersho)
        except:
            winery_short.append(' ')
        # 와이너리상세정보
        try:
            wineryin = bs.select_one('#section4 > div > div.box2 > dl > dd.txt').text
            winery_info.append(wineryin)
        except:
            winery_info.append(' ')

        # 이미지 저장
        # 이미지 링크 # '이미지 준비중입니다'
        img_url = bs.select_one('#section2 > div > div.left_box > img')['src']
        name_k = name_koli[-1].replace('\t', '')
        try:
            rq.urlretrieve(img_url, f'wine/{name_k}.jpg')
        except:
            url1, url2 = img_url.split('images/', maxsplit = 1)
            ur_ = url1 + 'images/' + quote(url2)
            rq.urlretrieve(ur_, f'wine/{name_k}.jpg')
```

## 1. 크롤링한 데이터 데이터프레임으로 생성

```python
# 데이터 만들기
elec_df = pd.DataFrame(columns = ['와인이름', '영어이름', '한줄소개', '와인타입', '나라, 와이너리', '포도종', '용량', '곁들일 음식', '당도', '산도', '바디', '와인설명', '수상', '와이너리이름', '와이너리설명'])
elec_df['와인이름'] = name_koli
elec_df['영어이름'] = name_enli
elec_df['한줄소개'] = short_int
elec_df['와인타입'] = winetype
elec_df['나라, 와이너리'] = country_winery
elec_df['포도종'] = grape_li
elec_df['용량'] = ml_li
elec_df['곁들일 음식'] = food_li
elec_df['당도'] = dang_li
elec_df['산도'] = acidity_li
elec_df['바디'] = body_li
elec_df['와인설명'] = infor_li
elec_df['수상'] = awards_li
elec_df['와이너리이름'] = winery_name
elec_df['와이너리특징'] = winery_short
elec_df['와이너리설명'] = winery_info
```
